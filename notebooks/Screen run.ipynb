# descending neuron activation bilateral

```python
simStr = "mancDnBilateralActivationScreen"
defaultParamPath = "../configs/default-manc-pulse.yaml"
useSeed = True

with open(defaultParamPath, 'r') as file:
    defaultParams = yaml.safe_load(file)
params = defaultParams
```

Change defaults

```python
# CHANGE DEFAULTS
wTable = pd.read_csv("../data/MANC weights/20231020_DNtoMN/wTable_20231020_DNtoMN_unsorted_withModules.csv",index_col=0) #TODO need better way of organizing which data version I'm using
allExcDNs = wTable.loc[(wTable["class"] == "descending neuron") & (wTable["predictedNt"] == "acetylcholine")]
excDnTypes = allExcDNs["type"].dropna().unique()

params["paramsToIterate"] = ["stimNeurons","seed","stimI"]

stimNeurons = []
seeds = []
stimI = []
for dnType in excDnTypes:
    dnTypeData = wTable.loc[wTable["type"] == dnType]

    stimNeurons.append(dnTypeData.index.to_list())
    seeds.append(int(np.random.default_rng().integers(10000)))
    stimI.append(float(np.mean((dnTypeData["size"]/512)/60000)))

nSims = len(stimNeurons)
params["stimNeurons"] = stimNeurons
params["seed"] = seeds
params["stimI"] = stimI
params["saveFigs"] = False

params
```

Save modified param file

```python
dateStr = date.today().strftime(format="%Y%m%d")
saveStr = f"{dateStr}_{simStr}"
fullConfigPath = f"../configs/{saveStr}.yaml"

with open(fullConfigPath,"w") as outfile:
    yaml.dump(params,outfile,sort_keys=False)
```

Run

```python
for i in range(nSims):
    os.system(f"tsp python {os.path.abspath('../src/run_from_config.py')} '{fullConfigPath}' --idx {i}")
```

```python
# RUN
for i in range(nSims):
    os.system(f"tsp python {os.path.abspath('../src/run_from_config.py')} '{fullConfigPath}' --idx {i}")
```

# interneuron deletion screen

```python
simStr = "mancBDN2InTypesSilencingScreen"
defaultParamPath = "../configs/default-manc-pulse.yaml"

with open(defaultParamPath, 'r') as file:
    defaultParams = yaml.safe_load(file)
params = defaultParams.copy()
```

First need to load the results of the activation screen and collect interneuron types that were active

```python
bdn2Config = "20240716_mancBDN2activation"
bdn2Dir = f"../results/tsp/{bdn2Config}/"
bdn2Dirs = os.listdir(bdn2Dir)
with open(f"../configs/{bdn2Config}.yaml", 'r') as file:
    bdn2Params = yaml.safe_load(file)
wTable = pd.read_csv(bdn2Params["dfPath"],index_col=0)
wTable.rename({"Unnamed: 0":"wIdx"},axis=1,inplace=True)
```

```python
allActiveIdxs = []
for dir in bdn2Dirs:
    simPath = bdn2Dir+dir
    if not os.path.isdir(simPath):
        bdn2Summary.drop(index=dir,inplace=True)
        continue

    R = np.load(simPath+"/R.npy")
    activeIdxs = np.where(np.sum(R,1)>0)[0]
    if len(activeIdxs)>500:
        # SKIP OVERACTIVE INDEXES
        print(f"Omitted {dir} due to overactivity")
        continue
    allActiveIdxs.extend(activeIdxs.tolist())

allActiveIdxs = np.unique(np.array(allActiveIdxs))
activeIns = wTable.loc[(wTable["wIdx"].isin(allActiveIdxs))&~(wTable["class"].isin(["descending neuron","motor neuron"]))]
activeInTypes = activeIns["type"].dropna().unique()
```

Change defaults

```python
stimNeurons = wTable.loc[wTable["type"] == "DNxl058"] # BDN2
params["stimNeurons"] = stimNeurons.index.to_list()
params["stimI"] = float(np.mean((stimNeurons["size"]/512)/60000))

params["paramsToIterate"] = ["removeNeurons","seed"]
removeNeurons = []
seeds = []
for inType in activeInTypes:
    removeNeurons.append(wTable.loc[wTable["type"] == inType].index.to_list())
    seeds.append(int(np.random.default_rng().integers(10000)))

params["removeNeurons"] = removeNeurons
params["seed"] = seeds
params["saveFigs"] = False

nSims = len(activeInTypes)

params
```

Save modified param file

```python
dateStr = date.today().strftime(format="%Y%m%d")
saveStr = f"{dateStr}_{simStr}"
fullConfigPath = f"../configs/{saveStr}.yaml"

with open(fullConfigPath,"w") as outfile:
    yaml.dump(params,outfile,sort_keys=False)
```

Run

```python
for i in range(nSims):
    os.system(f"tsp python {os.path.abspath('../src/run_from_config.py')} '{fullConfigPath}' --idx {i}")
```
